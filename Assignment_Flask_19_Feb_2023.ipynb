Abs 1 ) Flask is a web framework, it’s a Python module that lets you develop web applications easily. It’s has a small and easy-to-extend core: it’s a microframework that doesn’t include an 
       ORM (Object Relational Manager) or such features.
        It does have many cool features like url routing, template engine. It is a WSGI web app framework.
        Below are the advantages of flask framework:
        1. Scalable
        2. Flexible
        3. Easy to negotiate
        4. Lightweight
        5. Documentation

Ans 3) App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. 

Example: In our application, the URL (“/”) is associated with the root URL. So if our site’s domain was www.example.org and we want to add routing to “www.example.org/hello”, we would use “/hello”. 

To bind a function to an URL path we use the app.route decorator.

Ans 5 ) The url_for() method, is used to prepare a URL, for a function dynamically, such that, changing URLs, in the application, is avoided. 
       It accepts, the name of the view function, as the first argument, and, any number of keywords, to be sent(to the view function), as the second argument. 
Eg:

appFlask = Flask(__name__)
@appFlask.route('/home')

   return 'We are in Home Page!'

    print(url_for('login'))
