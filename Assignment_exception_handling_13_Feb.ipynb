Ans 3 ) Below are the errors defined in ArithmeticError class
--> ZeroDivisionError
--> OverflowError
--> FloatingPointError

```python
#ZeroDivisionErrror
try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
#OverflowError
j = 5.0
try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}") 
```
