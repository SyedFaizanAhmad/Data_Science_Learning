#Ans 2)
A scatter plot, is a chart that shows the relationship between two variables. 
They are an incredibly powerful chart type, allowing viewers to immediately understand a relationship or trend, which would be impossible to see in almost any other form.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python

plt.figure(figsize=(6,2))
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.title("This is scatter plot example")
plt.scatter(x,y)
```

# Ans 3)
In subplot() Function  the layout is organized in rows and columns, which are represented by the first and second argument. 
The third argument represents the index of the current plot. plt.subplot(2, 3, 1) #the answer figure has 2 rows, 3 columns, and this plot is the first plot.

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 3, 1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 3, 2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 3, 3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 3, 4)
plt.plot(x,y)
plt.show()
```

A barplot (or barchart) is one of the most common types of graphic. 
It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

#Ans 5 )
A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. 
In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median.
Matplotlib's boxplot mainly provides a graphical summary of a data set with features such as minimum, first quartile, median, third quartile, and maximum. 
Note: A quartile is a statistical phrase for dividing observations into four predetermined intervals based on data values

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
```

```python
fig = plt.figure(figsize =(10, 7))
# Creating plot
plt.boxplot(data) 
# show plot
plt.show()
```
