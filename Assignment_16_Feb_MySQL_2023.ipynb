Ans 4) DQL ( Data Query Language)
The commands of SQL that are used to retrieve data from the database are collectively called DQL. So all Select statements come under DQL.
The purpose of DQL Command is to get some schema relation based on the query passed to it.
Eg: SELECT *
      FROM table_name;
    Over here the select statement is the DQL.
    * denotes that all the columns from table named 'table_name' to be fetched.
    from is a keyword denoting which table to get data from.
    table_name will be replaced with actual table name to get data from.
    To choose specific columns from a table, mention the column names.
    Eg: Select emp_id,emp_name,emp_email from emp;

Ans 5 ) Primary Key: A primary key is used to ensure that data in the specific column is unique. A primary key column cannot have NULL values. 
When we define column as unique key column then in that case it may have NULL values.
It is either an existing table column or a column that is specifically generated by the database according to a defined sequence. 
Example: Table emp_detail has columns as emp_id, emp_name,emp_email,emp_doj.
Over here emp_id would be the primary key as this emp_id need to be unique.
CREATE TABLE emp_detail (
    emp_id int NOT NULL PRIMARY KEY,
    emp_name varchar(255) NOT NULL,
    emp_email varchar(255),
    emp_doj DATE NOT NULL
);
Foreign Key: 
A foreign key is a column or group of columns in a relational database table that provides a link between data in two tables. 
It is a column (or columns) that references a column (most often the primary key) of another table. 

Example: Table emp_course_enrolled(emp_id int FOREIGN KEY REFERENCES emp_detail(emp_id),
                                   course_id  int,
                                   enrolled_date DATE);
Over here the column emp_id is a foreign key referencing to the column emp_detail.emp_id.
Meaning that the emp_id in emp_course_enrolled cannot be other that what exists in emp_detail.emp_id.
Moreover, we cannot delete a parent record (in this case emp_detail) if it is being referenced in other table (over here emp_course_enrolled)

Ans 7) The order of execution of SQl clause is :
    Order	Clause	Function
1	FROM	Tables are joined to get the base data.

3	GROUP BY	The filtered base data is grouped.
4	HAVING	The grouped base data is filtered.
5	SELECT	The final data is returned.
6	ORDER BY	The final data is sorted.
